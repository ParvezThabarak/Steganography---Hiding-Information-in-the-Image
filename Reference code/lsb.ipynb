```python
#lsb-least significant bit
import numpy as np
```

```python
# ASCII CONVERSION
d = {chr(i): i for i in range(255)}  # character to ascii
c = {i: chr(i) for i in range(255)}  # ascii to character
```

```python
#message and encryption key
text = "Hello"
key = "key"
```

```python
#random pixel value image
x=np.random.randint(0,255,(10,10,3),dtype=np.uint8)
```

```python
x
```

```python
x_enc = x.copy()
n =0 # number of rows
m =0 # number of columns
z = 0 #colour panel
l = len(text)
kl=0
for i in range(l):
    char_val = d[text[i]]^d[key[kl]]
    for bit_pos in range(8):
        bit=(char_val  >> (7-bit_pos)) & 1 #important
        org_val = x_enc[n,m,z]
        x_enc[n, m, z] = np.uint8((org_val & 0xFE) | bit) #important
        print(f"Embedding bit {bit} of '{text[i]}' at ({n},{m},{z}) original={org_val} new={x_enc[n, m, z]}")
        z=(z+1)%3
        if z==0:
            m=m+1
            if m==x_enc.shape[1]:
                m=0
                n=n+1
    kl=(kl+1)%len(key)
```

```python
#decryption
n,m,z=0,0,0
k=0
decrypt=""
for i in range(l):
    val=0
    for bit_pos in range(8):
        bit= x_enc[n,m,z] & 1
        val=(val << 1) | bit
        print(f"Reading bit {bit} from ({n},{m},{z})")
        z=(z+1)%3
        if z==0:
            m=m+1
            if m==x_enc.shape[1]:
                m=0
                n=n+1
    orig_char= c[val ^ d[key[k]]]
    decrypt+=orig_char
    print(f"Decrypted byte: {val} XOR {d[key[k]]} = {val ^ d[key[k]]} -> '{orig_char}'")
    k=(k+1)%len(key)

print("Decrypt text", decrypt)
```
